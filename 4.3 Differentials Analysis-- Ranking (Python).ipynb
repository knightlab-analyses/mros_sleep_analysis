#### This notebook is to rank differentials for sleep measurements

```python
import pandas as pd
import numpy as np
```

```python
tax = pd.read_csv('taxonomy-table/taxonomy.tsv', sep='\t', header=0)
tax.head()
```

```python
df = pd.read_csv('Songbird/amplitute-table/metadata.tsv', sep='\t', header=0)
df = df.drop(df.index[0])
df = pd.merge(df, tax, right_on='Feature ID', left_on = 'featureid', how = 'left')
df = df.drop(columns = 'featureid')
cols = df.columns.drop(['Feature ID','Taxon'])
df[cols] = df[cols].apply(pd.to_numeric, errors='coerce')
df[cols] = df[cols].round(3) 
df = df.iloc[(-df['AMAMPT'].abs()).argsort()]
df = df[:10]
df = df[['Feature ID', 'Taxon', 'AMAMPT']]
df
```

```python
df.to_csv('amplitute.csv')
```
